To import masks as an array

```
images = []

for filename in mask:
  path = os.path.join('/content/gdrive/My Drive/good_gTruth/Masking/', filename)
  img = cv2.imread(path)
  if img is not None:
    images.append(img)

type(images)
print(images[0][0][0])
```

```
images = []
for filename in mask:
  path = os.path.join('/content/gdrive/My Drive/good_gTruth/Masking/', filename)
  img = cv2.imread(path)
  rows, cols = (224, 224) 
  one_img = [[0]*cols]*rows 
  for row in range(224):
    for col in range(224):
      if img[row][col][0] == 1:
        one_img[row][col] = 1
      elif img[row][col][0] == 2:
        one_img[row][col] = 2
      elif img[row][col][0] == 3:
        one_img[row][col] = 3
      elif img[row][col][0] == 4:
        one_img[row][col] = 4
      else:
        one_img[row][col] = 0

  images.append(one_img)
```

```
print(images.shape)
print(images[0].shape)
```

```
Y = np.zeros((len(mask), image_height, image_width), dtype = np.float32)
i=0
for image in images:
  Y[i] = image[i]
  i = i+1

print(type(Y))
print(Y.shape)
```

**FOR 7 BANDS**

```
from PIL import Image
import os
import numpy as np 
original = os.listdir('7_bands')
print(original)
X_7B = np.zeros((1554,1228, 7), dtype = np.float32)
i=0

for band in original:
  path = os.path.join('/content/gdrive/My Drive/good_gTruth/7_bands/', band)
  B = Image.open(path)
  X_7B[:,:,i]=B
  i=i+1
```

```
print(X_7B.size)
print(X_7B[0][0].size)
```

```
M_7B = np.zeros((1554,1228, 1), dtype = np.float32)
path = os.path.join('/content/gdrive/My Drive/good_gTruth/', 'Label_1.png')
B = Image.open(path)
M_7B[:,:,0]=B
```

```
print(M_7B.size)
print(M_7B[0][0].size)
```

```
X = np.zeros((250, 224, 224, 7), dtype = np.float32)
i=0
pixel_row=0
counter=1
while pixel_row < 1329: 
    pixel_col=0
    while pixel_col < 1005:
        pic_part = X_7B[pixel_row:(pixel_row+224), pixel_col:(pixel_col+224),:]
        X[i] = pic_part
        i = i+1
        pixel_col=pixel_col+111
        counter = counter+1
    pixel_row=pixel_row+55
```

```
print(X.shape)
```

```
Y = np.zeros((250, 224, 224, 1), dtype = np.float32)
i=0
pixel_row=0
counter=1
while pixel_row < 1329: 
    pixel_col=0
    while pixel_col < 1005:
        pic_part = M_7B[pixel_row:(pixel_row+224), pixel_col:(pixel_col+224),:]
        Y[i] = pic_part
        i = i+1
        pixel_col=pixel_col+111
        counter = counter+1
    pixel_row=pixel_row+55
```

```
print(Y.shape)
```

```
from sklearn.model_selection import train_test_split

X_tv, X_test, Y_tv, Y_test = train_test_split(X,Y,test_size=0.2,train_size=0.8)
X_train, X_val, Y_train, Y_val = train_test_split(X_tv,Y_tv,test_size = 0.20,train_size =0.80)
```

```
print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

```
print(Y_train.shape)
print(Y_val.shape)
print(Y_test.shape)
```

```
from segmentation_models import get_preprocessing
from segmentation_models.losses import bce_jaccard_loss
from segmentation_models.losses import CategoricalCELoss
from segmentation_models.metrics import iou_score
from segmentation_models.metrics import FScore
from segmentation_models.metrics import Precision
from segmentation_models.metrics import Recall
from segmentation_models import Unet
```

```
from keras.layers import Input, Conv2D
from keras.models import Model


N = X_train.shape[-1]
loss = CategoricalCELoss()
metric = FScore()
p = Precision()
r = Recall()

base_model = Unet(backbone_name='resnet34', encoder_weights='imagenet')

inp = Input(shape=(224, 224, 7))
l1 = Conv2D(3, (1, 1))(inp) # map N channels data to 3 channels
out = base_model(l1)

model = Model(inp, out, name=base_model.name)
model.compile('Adam', loss=loss, metrics=['accuracy',metric, p, r, iou_score])

history = model.fit(
    x=X_train,
    y=Y_train,
    batch_size=16,
    epochs=100,
    validation_data=(X_val, Y_val),
)
```
